```python
import pandas as pd
import os
from functools import reduce
```

```python
tabelas = []

for i in range(5, 22):

    todas_tabelas = f"tabela_{i}.csv"

    try:

        df = pd.read_csv(todas_tabelas, encoding="utf-8")
    except UnicodeDecodeError:
        df = pd.read_csv(todas_tabelas, encoding="latin1")
        
    tabelas.append(df)
    display(df)
```

```python
colunas_chave = ['UF', 'IBGE', 'Nome do MunicÃ­pio', 'Ãndice de Vulnerabilidade Social', 'NÃºmero de habitantes segundo o IBGE 2022', 'Faixa de Porte Populacional segundo o IBGE 2022', 'Ãndice de Equidade e Dimensionamento (IED)']
```

```python
for i, df in enumerate(tabelas):
    colunas_faltantes = [col for col in colunas_chave if col not in df.columns]
    if colunas_faltantes:
        print(f"⚠️ Tabela {i+5} ({todas_tabelas}) não tem as colunas: {colunas_faltantes}")

```

```python
tabelas_validas = [df for df in tabelas if all(col in df.columns for col in colunas_chave)]

```

```python
tabela_mesclada = reduce(lambda left, right: pd.merge(left, right, on=colunas_chave, how="outer"), tabelas)
```

```python
tabela_mesclada.to_csv("tabelas_mescladas.csv", encoding="utf-8", index=False)
```
