```python
# Lista de arquivos CSV
arquivos = [
    'dados/Mortalidade_Geral_2018.csv',
    'dados/Mortalidade_Geral_2019.csv',
    'dados/Mortalidade_Geral_2020.csv',
    'dados/Mortalidade_Geral_2021.csv',
    'dados/DO22OPEN.csv',
    'dados/DO23OPEN.csv'
]

# Lista para armazenar os DataFrames
dataframes = []

# Identificar colunas padrão (a partir do primeiro arquivo)
colunas_referencia = pd.read_csv(arquivos[0], sep=';', nrows=1).columns

# Loop para ler e ajustar os arquivos
for arquivo in arquivos:
    try:
        # Carregar o arquivo
        df = pd.read_csv(arquivo, sep=';', low_memory=False)
        
        # Verificar se o número de colunas está correto
        if df.shape[1] < len(colunas_referencia):
            print(f'Ajustando colunas do arquivo: {arquivo}')
            
            # Adicionar coluna faltante com valores NaN
            for col in colunas_referencia:
                if col not in df.columns:
                    df[col] = None  # Preencher com valores vazios

        # Reordenar colunas para manter consistência
        df = df[colunas_referencia]

        # Adicionar DataFrame à lista
        dataframes.append(df)

    except Exception as e:
        print(f'Erro ao processar {arquivo}: {e}')

# Concatenar todos os DataFrames
df_final = pd.concat(dataframes, ignore_index=True)

# Exibir resumo final
print(f'Total de Linhas: {df_final.shape[0]}')
print(f'Total de Colunas: {df_final.shape[1]}')

# Salvar arquivo concatenado
df_final.to_csv('dados/Mortalidade_Geral_Completo.csv', sep=';', index=False)
print('Arquivo final salvo como dados/Mortalidade_Geral_Completo.csv')
```

```python
def analisar_relacao_obitos_cid(caminho_arquivo):
    # Carregar a planilha
    df = pd.read_csv(caminho_arquivo, sep=';', encoding='utf-8')
    
    # Verificar as primeiras linhas
    print("Visualizando as primeiras linhas do arquivo:")
    print(df.head())
    
    # Ajustar nomes de colunas, se necessário (exemplo genérico)
    colunas_esperadas = ['CAUSABAS', 'TIPOBITO']
    df = df[colunas_esperadas]
    
    # Garantir que TIPOBITO seja numérico
    df['TIPOBITO'] = pd.to_numeric(df['TIPOBITO'], errors='coerce').fillna(0)
    
    # Agrupar os óbitos por CID
    df_grouped = df.groupby('CAUSABAS')['TIPOBITO'].sum().reset_index()
    
    # Verificar dados após agrupamento
    print(df_grouped.head())
    print(df_grouped.describe())
    
    # Criar um índice categórico para CAUSABAS
    df_grouped['CAUSABAS_Index'] = df_grouped['CAUSABAS'].astype('category').cat.codes
    
    # Análise estatística (exemplo: Teste de Correlação de Spearman)
    correlacao, p_valor = stats.spearmanr(df_grouped['CAUSABAS_Index'], df_grouped['TIPOBITO'])
    
    print(f"Correlação de Spearman: {correlacao:.3f}, p-valor: {p_valor:.3f}")
    
    # Visualizar relação
    plt.figure(figsize=(12, 6))
    top_causas = df_grouped.nlargest(20, 'TIPOBITO')  # Mostrar apenas os 20 principais
    sns.barplot(data=top_causas, x='CAUSABAS', y='TIPOBITO', palette='viridis')
    plt.xticks(rotation=90)
    plt.xlabel('CAUSABAS')
    plt.ylabel('Número de Óbitos')
    plt.title('Relação entre CAUSABAS e TIPOBITO')
    plt.show()
    
    # Interpretar resultado
    if p_valor < 0.05:
        print("Existe uma relação estatisticamente significativa entre CAUSABAS e TIPOBITO.")
    else:
        print("Não foi encontrada uma relação estatisticamente significativa entre CAUSABAS e TIPOBITO.")



analisar_relacao_obitos_cid('dados/Mortalidade_Geral_Completo.csv')

```
